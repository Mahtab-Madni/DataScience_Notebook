# Key Variations of ARIMA Models :

#### **ARIMA (p, d, q) :**  The standard non-seasonal model combining Autoregressive (AR) terms (p), Differencing (I) to remove trends (d), and Moving Average (MA) errors (q).

#### **SARIMA (Seasonal ARIMA):** Denoted as (ARIMA(p,d,q) * (P,D,Q)s), this model extends ARIMA to handle seasonal patterns (s) by adding seasonal components (P,D,Q).
#### **SARIMAX (Seasonal ARIMA with eXogenous variables):** A SARIMA model that includes additional external data (covariates or independent variables) to improve forecasting accuracy.
#### **ARIMA with Exogenous Variables (ARIMAX):** A non-seasonal model that incorporates external regressors
#### **VARMA (Vector Autoregressive Moving Average):** An extension used for multivariate time series forecasting, where multiple time series depend on each other.Â

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

data = pd.read_csv('DataSet\\date_count.csv')

data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
data = data.asfreq('D')
data = data.bfill()
```

```python
plt.figure(figsize=(12, 5))
plt.plot(data, label='Count')
plt.title('Daily Count Over Time')
plt.legend()
plt.show()
```

```python
decomposition = seasonal_decompose(data['count'], model='additive')
decomposition.plot()
plt.show()
```

```python
train = data.iloc[:-30]
test = data.iloc[-30:]
```

```python
model = SARIMAX(train['count'], order=(1,1,1), seasonal_order=(1,1,1,7))
result = model.fit(disp=False)

forecast = result.predict(start=test.index[0], end=test.index[-1], dynamic=False)

mse = mean_squared_error(test['count'], forecast)
mae = mean_absolute_error(test['count'], forecast)
r2 = r2_score(test['count'], forecast)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"R2 Score: {r2:.2f}")

plt.figure(figsize=(14, 6))
plt.plot(train.index, train['count'], label='Train')
plt.plot(test.index, test['count'], label='Real')
plt.plot(forecast.index, forecast, label='Forecast')
plt.title('Actual vs Forecasted Values')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.statespace.sarimax import SARIMAX

data = pd.read_csv('DataSet/date_count.csv', parse_dates=['Date'])
data = data.set_index('Date')

order = (1, 1, 1)
seasonal_order = (1, 1, 1, 7)

model = SARIMAX(data['count'], 
                order=order,
                seasonal_order=seasonal_order,
                freq='D')
results = model.fit(disp=False)

forecast = results.get_forecast(steps=30)
predicted_values = forecast.predicted_mean
confidence_intervals = forecast.conf_int()

plt.figure(figsize=(14, 7))

plt.plot(data.index, data['count'], label='History Data', color='blue')

pred_index = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=30)
plt.plot(pred_index, predicted_values, label='Forecast', color='red')

plt.fill_between(pred_index,
                 confidence_intervals.iloc[:, 0],
                 confidence_intervals.iloc[:, 1],
                 color='pink', alpha=0.3, label='95% Confidence Interval')

plt.title('SARIMA Forecast with Confidence Interval', fontsize=14)
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(results.summary())
```

```python
from statsmodels.tsa.arima.model import ARIMA 

# 1. Check if there really is no seasonality (via decomposition)
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(data['count'], model='additive', period=7)  # Test with suspected period
decomposition.plot()
plt.show()

# 2. ARIMA modeling (only non-seasonal components)
order = (1, 1, 1)  # (p, d, q)
model = ARIMA(data['count'], order=order)
results = model.fit()

# 3. Forecast
forecast = results.get_forecast(steps=30)
predicted_values = forecast.predicted_mean

# 4. Plot (same as before, but without seasonal component)
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['count'], label='Actual Data')
plt.plot(predicted_values.index, predicted_values, color='red', label='ARIMA Forecast')
plt.legend()
plt.show()
```

```python
from prophet import Prophet

# 1. Prepare the data
data = pd.read_csv('DataSet/date_count.csv')
df = data.rename(columns={'Date': 'ds', 'count': 'y'})

# 2. Create and train the model
model = Prophet(
    yearly_seasonality=False,  # Disable yearly seasonality
    weekly_seasonality=True,   # Enable weekly seasonality
    daily_seasonality=False    # Disable daily seasonality
)
model.fit(df)  # Now it will work!

# 3. Make forecasts
future = model.make_future_dataframe(periods=30, freq='D')
forecast = model.predict(future)

# 4. Visualize results
fig1 = model.plot(forecast)
fig2 = model.plot_components(forecast)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# 1. Separate data (last 30 days for testing)
train = df[:-30]
test = df[-30:]

# 2. Train with data up to the cutoff date
model = Prophet()
model.fit(train)

# 3. Create future dataframe only for the test period
future = model.make_future_dataframe(periods=30, freq='D')
forecast = model.predict(future)

# 4. Filter only the predictions for the test period
predictions = forecast.tail(30)['yhat']

# 5. Calculate metrics
mae = mean_absolute_error(test['y'], predictions)
rmse = np.sqrt(mean_squared_error(test['y'], predictions))

print(f"MAE: {mae:.2f}, RMSE: {rmse:.2f}")
```

```python
from prophet.diagnostics import cross_validation

# Time-based cross-validation
df_cv = cross_validation(
    model,
    initial='500 days',  # Initial training period
    period='30 days',    # Interval between cutoffs
    horizon='30 days'    # Size of each test period
)

# Calculate aggregated metrics
from prophet.diagnostics import performance_metrics
df_perf = performance_metrics(df_cv)
print(df_perf.head())
```
